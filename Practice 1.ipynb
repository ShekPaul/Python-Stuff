```python
# s= "Hello"http://localhost:8888/notebooks/Desktop/Stuff/Stuff.ipynb#Reversing-the-String-Methods
def reverse_(s):
    ss = "".join(reversed(s))
    return ss

reverse_(string)
```

```python
def reverse(s):
    str = ""
    for i in s:
        str = i + str
    return str

reverse(string)
```

```python
def rever(s):
    s= s[::-1]
    return s
    
rever(string)
```
