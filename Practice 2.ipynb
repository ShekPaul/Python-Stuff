```python
def multi(n):
    numbers= []
    for i in range(n):
        if (i%3==0) or (i%5==0):
            numbers.append(i)
    print(numbers)
        
multi(20)
```

```python
# prime number

n=int(input("Enter the number: "))
while n>1:
    for i in range(2,int(n/2)+1):
        if n%i==0:
            print(n,'is not a Prime number')
            break
    else:
        print(n, 'is a Prime Number')
        
    
else:
    print('Not a Prime number. \nEnter number greater than 2')
```
